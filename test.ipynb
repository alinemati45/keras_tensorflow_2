```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
```

```python
epochs = 200
bactch_size = 128
verbose = 1
nb_classe = 10
validation = 0.2
n_hide = 128
mnsit = keras.datasets.mnist
(X_train , Y_train) , (X_test , Y_test) = mnsit.load_data()
X_train.shape 
```

```python
reshape = 28 * 28
X_train = X_train.reshape(60000 ,reshape )
X_test = X_test.reshape(10000 , reshape)
X_train
```

```python
X_train.dtype 
```

```python
X_train = X_train.astype("float32")
X_test = X_test.astype("float32")
```

```python
np.max(X_train) 
```

```python
X_train /= 255
X_test /= 255
print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')
```

```python
Y_train = tf.keras.utils.to_categorical(Y_train , nb_classe)
Y_test = tf.keras.utils.to_categorical(Y_test , nb_classe)
```

```python
model = tf.keras.models.Sequential()
model.add(keras.layers.Dense(nb_classe , input_shape = (reshape ,) , name="dense_layer" , activation="softmax"))
```

```python
model.summary()
```

```python
model.compile(opimizer ='SDG' , loss='categorical_crossentropy' , metrics=["accuracy"])
```

```python
model.fit(X_train , Y_train ,batch_size = bactch_size , epochs = epochs  , verbose = 0 , validation_split = validation)

test_loss , test_acc = model.evaluate(X_test , Y_test)
```
